# Summary

This is a notebook for the "Housing Prices Competition for Kaggle Learn Users" Competition.

Note: This notebook will not discuss business aspects and data-driven decision making. This is an extremely important aspect in the field of data science, but it will be discussed at another time.

The models used were:
- Lasso 
- Random Forest 
- Decision Tree 
- AdaBoost 
- Gradient Boosting 
- Extreme Gradient Boosting 

The basic pipeline was same of last notebook composite by:
- Drop features with more than 40% of missing values. 
- Impute missing values with mean for numerical columns and most_frequent for categorical columns.
- Transform numerical features by scaling with MinMaxScaler()
- Encode categorical features with OneHotEncoding 
 
The best Mean Absolute Error (MAE) score was obtained by Gradient Boosting with a value of 15478.32 (Top 5%). 

Just choosing the right model made the score jump from the top 62 to the top 5%!

```python
import numpy as np  
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt 
```

```python
df_train = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
print(df_train.shape)
df_train.head(5)
```

```python
print(df_test.shape)
df_test.head(5)
```

```python
df_train.info()
```

```python
df_train.describe().T
```

```python
df_train.describe(include=np.object).T
```

```python
train_missing_prop = df_train.loc[:,df_train.isna().sum()>0].isna().mean(axis=0)

fig = plt.figure(figsize=(12,6))
train_missing_prop.plot.bar()
plt.show()
```

```python
test_missing_prop = df_test.loc[:,df_test.isna().sum()>0].isna().mean(axis=0)

fig = plt.figure(figsize=(12,6))
test_missing_prop.plot.bar()
plt.show()
```

```python
threshold = 0.4
train_over_threshold = train_missing_prop[train_missing_prop >= threshold]
test_over_threshold = test_missing_prop[test_missing_prop >= threshold]

print(train_over_threshold)
print('')
print(test_over_threshold)
```

```python
cols_drop = train_over_threshold.index.tolist()
df_train_drop = df_train.drop(cols_drop, axis=1)
df_test_drop = df_test.drop(cols_drop, axis=1)
```

## Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_train_drop.drop(['Id', 'SalePrice'], axis=1)
y = df_train_drop.SalePrice
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=0)
```

```python
X_test = df_test_drop.drop('Id', axis=1)
```

```python
print(X_train.shape, X_valid.shape, X_test.shape)
```

## Pipeline

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
numerical_features = X_train.select_dtypes(include='number').columns.tolist()
print(len(numerical_features))
print(numerical_features)
```

```python
categorical_features = X_train.select_dtypes(exclude='number').columns.tolist()
print(len(categorical_features))
print(categorical_features)
```

```python
#mudei mean by median (reverter depois)
numeric_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='median')),
    ('scale', MinMaxScaler())
])

categorical_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('one-hot', OneHotEncoder(handle_unknown='ignore', sparse=False))
])
```

### Column Transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
full_processor = ColumnTransformer(transformers=[
    ('number', numeric_pipeline, numerical_features),
    ('category', categorical_pipeline, categorical_features)
])
```

### Final Pipeline With an Estimator

```python
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor

from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import RepeatedKFold
```

```python
cv = RepeatedKFold(n_splits=5, n_repeats=2, random_state=0)
```

```python
def Randomized_Search(model, param_space, cv):
    search = RandomizedSearchCV(model, param_space, cv=cv, n_iter=10,
                               scoring='neg_mean_absolute_error', random_state=0)

    _ = search.fit(X_train, y_train)
    
    print('Best parameters:', search.best_params_)
    print('Best score:', abs(search.best_score_))
    
    return search.best_estimator_
```

```python
lasso = Lasso(max_iter=2000, tol=1e-2)

lasso_pipeline = Pipeline(steps=[
    ('preprocess', full_processor),
    ('model', lasso)
])

param_space = {'model__alpha':np.linspace(0, 4, 40)}

best_lasso = Randomized_Search(lasso_pipeline, param_space, cv)
```

```python
rf = RandomForestRegressor()

rf_pipeline = Pipeline(steps=[
    ('preprocess', full_processor),
    ('model', rf)
])

param_space = {'model__n_estimators': np.arange(20,800,20)}

best_rf = Randomized_Search(rf_pipeline, param_space, cv)
```

```python
dt = DecisionTreeRegressor(random_state = 0)

dt_pipeline = Pipeline(steps=[
    ('preprocess', full_processor),
    ('model', dt)
])

param_space = {'model__max_depth': np.arange(2, 14,2)}

best_dt = Randomized_Search(dt_pipeline, param_space, cv)
```

```python
dt = DecisionTreeRegressor(random_state = 0, max_features = "auto", max_depth = None)

adaB = AdaBoostRegressor(base_estimator = dt, random_state=0)

adaB_pipeline = Pipeline(steps=[
    ('preprocess', full_processor),
    ('model', adaB)
])

param_space = { 'model__n_estimators': np.arange(20,800,20)}

best_adaB = Randomized_Search(adaB_pipeline, param_space, cv)
```

```python
gb = GradientBoostingRegressor(random_state=0)

gb_pipeline = Pipeline(steps=[
    ('preprocess', full_processor),
    ('model', gb)
])

param_space = {'model__n_estimators':  np.arange(20,600,20)}

best_gb = Randomized_Search(gb_pipeline, param_space, cv)
```

```python
xgb = XGBRegressor(random_state=0)

xgb_pipeline = Pipeline(steps=[
    ('preprocess', full_processor),
    ('model', xgb)
])

param_space = { 'model__n_estimators': np.arange(10,300,10),
                'model__learning_rate': np.linspace(0.01, 1, 10)}

best_xgb = Randomized_Search(xgb_pipeline, param_space, cv)
```

### Prediction on validation set

```python
MAE = []
models = [best_lasso, best_rf, best_dt, best_adaB, best_gb, best_xgb]
ind = ['Lasso', 'RForest', 'DTree', 'AdaB', 'GB', 'XGB']

for i in models:
    pred = i.predict(X_valid)
    MAE.append(mean_absolute_error(y_valid, pred))
 
```

```python
df_mae = pd.DataFrame(MAE, index = ind, columns=['MAE'])
df_mae.sort_values('MAE', ascending=True, inplace=True)
df_mae
```

```python
df_mae.plot.bar(legend=False)
plt.ylabel('MAE')
plt.xticks(rotation=0)
plt.show()
```

### Prediction on test set

```python
pred_test_gb = best_gb.predict(X_test)
pred_test_xgb = best_xgb.predict(X_test)
```

```python
output_gb = pd.DataFrame({'Id': df_test.Id,
                       'SalePrice': pred_test_gb})

output_xgb = pd.DataFrame({'Id': df_test.Id,
                       'SalePrice': pred_test_xgb})
```

```python
output_gb.to_csv('submissions/submission2_gb.csv', index=False)
output_xgb.to_csv('submissions/submission2_xgb.csv', index=False)
```
